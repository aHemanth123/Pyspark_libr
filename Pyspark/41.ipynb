```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

data =[(1,'mahe'),(2,'wafa')]
print(type(data))
```

TO Convert list to Rdd

```python
rd=spark.sparkContext.parallelize(data)
print(type(rd))

print(rd.collect())  # collecting the information
```

Rdd to Dataframe

```python
# # u need to convert into df  

# 1
df = rd.toDF(schema =['i d', 'name'])
#2
# df  = spark.createDataFrame(rd,['id','name'])

df.printSchema()
```
