```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

data =[(1,'mahe'),(2,'wafa')]

rd=spark.sparkContext.parallelize(data)
print(rd.collect())

# it is not  working  new rdd
rdd10 = rd.map(lambda x: x +(x[0]+ ' '+x[1] ,))   # tuple like , # map  function is used to combine x[0] +x[1]

print(rdd10.collect())

```

------------------------
Create a Dataframe   and  generate Rdd from it 

```python
df = spark.createDataFrame(data,['id','name'])

rdd1 = df.rd.map(lambda x: x +(x[0]+ ' '+x[1] ,)) #------------------------
# Create a Dataframe   and  generate Rdd from it 

df2 =rdd1.toDF(['id','name'])
df2.printSchema()
```
